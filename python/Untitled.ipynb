```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
G.add_node(r'$x_1$')
G.add_node(r'$x_2$')
G.add_node(r'$x_3$')
G.add_node(r'$x_4$')
G.add_node(r'$x_5$')
G.add_node(r'$x_6$')
G.add_node(r'$x_7$')
G.add_node(r'$\beta_1$')
G.add_node(r'$\beta_2$')
G.add_node(r'$\beta_3$')
G.add_node(r'$\beta_4$')
G.add_node(r'$\beta_5$')
G.add_node(r'$\beta_6$')
G.add_node(r'$\beta_7$')
G.add_node(r'$\beta_8$')
G.add_node(r'$\lambda_{y_{i}}$')
G.add_node(r'$y_{i}$')
G.add_node(r'$\mu_{\beta}$')
G.add_node(r'$\sigma_{\beta}$')
```

```python
G.add_edge(r'$x_1$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_2$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_3$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_4$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_5$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_6$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$x_7$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_1$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_2$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_3$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_4$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_5$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_6$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_7$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\beta_8$', r'$\lambda_{y_{i}}$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_1$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_2$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_3$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_4$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_5$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_6$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_7$')
G.add_edge(r'$\mu_{\beta}$', r'$\beta_8$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_1$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_2$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_3$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_4$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_5$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_6$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_7$')
G.add_edge(r'$\sigma_{\beta}$', r'$\beta_8$')
G.add_edge(r'$\lambda_{y_{i}}$', r'$y_{i}$')
```

```python
layout = {
    r'$x_1$': (-30, 15),
    r'$x_2$': (-30, 10),
    r'$x_3$': (-30, 5),
    r'$x_4$': (-30, 0),
    r'$x_5$': (-30, -5),
    r'$x_6$': (-30, -10),
    r'$x_7$': (-30, -15),
    r'$\beta_1$': (-15, 30),
    r'$\beta_2$': (-10, 30),
    r'$\beta_3$': (-5, 30),
    r'$\beta_4$': (0, 30),
    r'$\beta_5$': (5, 30),
    r'$\beta_6$': (10, 30),
    r'$\beta_7$': (15, 30),
    r'$\beta_8$': (20, 30),
    r'$\lambda_{y_{i}}$': (0, 0),
    r'$y_{i}$': (20, 0),
    r'$\mu_{\beta}$': (-10, 45),
    r'$\sigma_{\beta}$': (10, 45),
}
```

```python
# Need to create a layout when doing
# separate calls to draw nodes and edges
plt.figure(figsize=(20,20))
pos = nx.spectral_layout(G)
nx.draw_networkx_nodes(G, layout, node_size=5000)
nx.draw_networkx_labels(G, layout, font_size=25)
nx.draw_networkx_edges(G, layout, arrows=True)
plt.show()
```

```python
H = nx.moral_graph(G)
```

```python
layout = nx.circular_layout(H)
plt.figure(figsize=(20,20))
nx.draw_networkx_nodes(H, layout, node_size=2000)
nx.draw_networkx_labels(H, layout, font_size=25)
nx.draw_networkx_edges(H, layout, connectionstyle='arc3,rad=5')
plt.show()
```
